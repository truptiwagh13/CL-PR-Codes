```python
import nltk
```

```python
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
```

```python
sen="Hello Trupti i am a computer Science Engineering Student"
```

```python
token=word_tokenize(sen)
```

```python
token
```

```python
for i in token:
    print(i)
```
